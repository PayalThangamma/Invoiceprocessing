```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import pytesseract
import re
```

```python
def plot_gray(image):
    plt.figure(figsize=(15,10))
    return plt.imshow(image, cmap='Greys_r')
```

```python
def plot_rgb(image):
    plt.figure(figsize=(15,10))
    return plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
```

```python
import cv2
import aspose.words as aw
import numpy as np
import pytesseract
from pdf2image import convert_from_path
import os, pathlib
from os import listdir
```

```python
pytesseract.pytesseract.tesseract_cmd = "C:\\Program Files (x86)\\Tesseract-OCR\\tesseract.exe"
directory = 'C:\\Users\\payal\\OneDrive\\Desktop\\dataset\\images\\invoice'
```

```python
i=1
for filename in os.listdir(directory):
    if filename.endswith('.pdf'):
        with open(os.path.join(directory, filename)) as f:
            images = convert_from_path(pathlib.Path(filename), 100)
            for image in images:
                image.save("C:\\Users\\payal\\OneDrive\\Desktop\\dataset\\images\\invoice\\new\\output-%i.png" % i)
                i=i+1
```

```python
img_dir = "C:\\Users\\payal\\OneDrive\\Desktop\\dataset\\images\\invoice\\new"
```

```python
j=1
for images in os.listdir(img_dir):
    img = cv2.imread(os.path.join(img_dir, images))
    text = pytesseract.image_to_string(img)
    with open("C:\\Users\\payal\\OneDrive\\Desktop\\dataset\\images\\invoice\\out-%i.txt" % j, 'w', encoding="utf-8") as f:
        f.write(text)
    j=j+1 
```
